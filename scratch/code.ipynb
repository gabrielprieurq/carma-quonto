```python
def inverse(mot):
    c=""
    for i in range(len(mot)):
        c+=mot[len(mot)-1-i]
    return c

inverse("abcd")
```

```python
def estdanschaine(petit,grand):
    i=0
    c=0
    while i<len(grand):
        if petit[c]==grand[i]:
            c+=1
            if c==len(petit):
                return True
        else:
            c=0
        i+=1
    return False
estdanschaine("add","abdc")
```

```python
def factorielle(n):
    if n==1:
        return 1
    else:
        return n*factorielle(n-1)
```
