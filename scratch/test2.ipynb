```python
import numpy as np

def nombrepremier(n):
    if n==1:
        return False
    for i in range (2,int(np.sqrt(n))+1):
        if n%i==0:
            return False
    return True
```

```python
nombrepremier(4)
```

```python

def generepremier(p):
    L=[]
    c=0
    i=1
    while c<p:
        if nombrepremier(i):
            L.append(i)
            c+=1
        i+=1
    return L
generepremier(10)
```
